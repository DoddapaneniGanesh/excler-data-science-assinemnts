# Task 2: Dimensionality Reduction with PCA:

```python
# Standardized the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# Implementing PCA
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(scaled_data)
```

```python
# Determining the optimal number of principal components
import numpy as np
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
```

```python
# Scree plot to visualize the explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()
```

```python
#Transforming  the original dataset into the principal components

n_components = 3
pca = PCA(n_components=n_components)
pca_data = pca.fit_transform(scaled_data)
```

# Task 3: Clustering with Original Data:

```python
# K-means clustering with 3 clusters

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
# Adding cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Alcohol'], df['Malic'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Alcohol')
plt.ylabel('Malic')
plt.title('K-means Clustering with 3 Clusters')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score = silhouette_score(df, df['Cluster'])
print('Silhouette score:', silhouette_score)
```

```python
from sklearn.metrics import davies_bouldin_score

davies_bouldin_score = davies_bouldin_score(df, df['Cluster'])
print('Daviesâ€“Bouldin index:', davies_bouldin_score)
```

# Task 4: Clustering with PCA Data:

```python
#  Applying  the same clustering algorithm to the PCA-transformed dataset.

# K-means clustering with 3 clusters on PCA-transformed data 

kmeans = KMeans(n_clusters=3)
kmeans.fit(pca_data)

# Adding the  cluster labels to the DataFrame here 

df['Cluster_PCA'] = kmeans.labels_

```

```python
plt.figure(figsize=(10, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=df['Cluster_PCA'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering with 3 Clusters on PCA-transformed data')
plt.show()
```

```python
print('Clustering results on original data:')
print(df.groupby('Cluster').mean())

print('Clustering results on PCA-transformed data:')
print(df.groupby('Cluster_PCA').mean())
```
