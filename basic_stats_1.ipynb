# Standardization of Numerical Variables

# Explain the concept of standardization (z-score normalization).
Standardization (z-score normalization) is a technique used to transform numerical variables to have a mean of 0 and a standard deviation of 1. This makes the variables more comparable and helps to improve the performance of many analytical models."

# 	Standardize the numerical columns using the formula: z=x-mu/sigma


```python
# Standardize the numerical columns using the formula: z=x-mu/sigma
for col in numerical_columns:
    df[f"zscore_{col}"] = (df[col] - df[col].mean()) / df[col].std()
```

# 	Show before and after comparisons of the data distributions.

```python
# Show before and after comparisons of the data distributions.
for col in numerical_columns:
    print(f"Before standardization:")
    print(df[col].describe())
    print(f"After standardization:")
    print(df[f"zscore_{col}"].describe())  
```
